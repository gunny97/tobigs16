```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Mall_Customers.csv')
```

```python
data.head()
```

```python
# feature에 상관 없는 customer id drop
data.drop('CustomerID',axis=1,inplace=True)
```

```python
data.describe()
```

```python
sns.countplot(data['Gender'])
```

```python
for col in data.columns:
    if data[col].dtype == 'int64':
        sns.distplot(data[col])
        plt.show()

    else:
        continue
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
dummy_data = pd.get_dummies(data['Gender'])
data.drop('Gender',axis=1,inplace=True)
data = pd.concat([data,dummy_data],axis=1)
data.head()
```

## PCA

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
data = scaler.fit_transform(data)


# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 2) 
data = pca.fit_transform(data)
```

## Clustering

```python
plt.figure(figsize=(10,7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(data[:,0],data[:,1],label='True Position')
labels = [i for i in range(len(data))]
for label, x, y in zip(labels, data[:, 0], data[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

### Hierarchical clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(data, 'weighted')

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labels,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

### K-means clustering

```python
from sklearn.cluster import KMeans

cluster_candidates = [i+1 for i in range(10)]
inertia_list = []

for i in cluster_candidates:
    kmeans_model = KMeans(n_clusters =i, init = 'k-means++',random_state=123)
    kmeans_model.fit(data)
    inertia_list.append(kmeans_model.inertia_)
    
plt.plot(range(1, 11), inertia_list)
plt.title('The Elbow Method', fontsize='23')
plt.xlabel('# of clusters')
plt.xticks(np.arange(1,11))
plt.ylabel('inertia')
plt.show()
```


#### Elbow Method에 따르면 4번째 cluster에서 inertia값이 크게 바뀜 -> 최적의 클러스터 갯수 설정

```python
from sklearn.metrics import silhouette_score
import random

cluster_candidates = [3,4,5,6,7]

for cluster in cluster_candidates:
    num_cluster = cluster
    kmeans_cluster = KMeans(n_clusters=num_cluster, init='k-means++', random_state=123)
    clusters = kmeans_cluster.fit_predict(data)
    
    score = silhouette_score(data, clusters)
    print("For n_clusters = {}, silhouette score is {})".format(num_cluster, score))


    for i in range(num_cluster):
        r = random.random()
        b = random.random()
        g = random.random()
        color = (r, g, b)
        plt.scatter(data[clusters == i, 0], data[clusters == i, 1], 
                    s = 80, c = color, label = f'Cluster{i+1}')

    plt.scatter(kmeans_cluster.cluster_centers_[:, 0], kmeans_cluster.cluster_centers_[:, 1], s = 150, c = 'red', label = 'Centroids')
    plt.show()
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
import mglearn


db_cluster = DBSCAN(eps=0.3,min_samples=3)
clusters = db_cluster.fit_predict(data)

plt.scatter(data[:, 0], data[:, 1], c=clusters, s=60, edgecolors='black')
```

```python
import hdbscan

hdb__clusterer = hdbscan.HDBSCAN(min_cluster_size=15, gen_min_span_tree=True)
hdb__clusterer.fit(data)

palette = sns.color_palette()
cluster_colors = [sns.desaturate(palette[col], sat)
                  if col >= 0 else (0.3,0.3,0.3) for col, sat in
                  zip(hdb__clusterer.labels_, hdb__clusterer.probabilities_)]
plot_kwds = {'alpha' : 0.25, 's' : 80, 'linewidths':0}
plt.scatter(data[:, 0], data[:, 1], c=cluster_colors, **plot_kwds)
```

### Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

cluster_candidates = [3,4,5,6,7]

for cluster in cluster_candidates:
    agg_clu = AgglomerativeClustering(n_clusters=cluster, affinity='euclidean', linkage='ward')
    pred = agg_clu.fit_predict(data)
    
    score = silhouette_score(data, pred)
    print("For n_clusters = {}, silhouette score is {})".format(cluster, score))
    
    plt.figure(figsize=(10, 7))
    plt.scatter(data[:,0], data[:,1], c=agg_clu.labels_, cmap='rainbow')
    plt.show()
```
